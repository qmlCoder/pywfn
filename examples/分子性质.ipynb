```python
angs=np.linspace(0,360,19)[:-1]
r=4.15
cords=[]
for ang in angs:
    x=np.cos(ang/180*np.pi)*r
    y=np.sin(ang/180*np.pi)*r
    cords.append([x,y,0])
cords=np.array(cords)
for x,y,z in cords:
    print(f"C      {x:>16.8f} {y:>16.8f} {z:>16.8f}")
```
