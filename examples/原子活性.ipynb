```python
import sys
sys.path.append(rf"D:\code\pywfn")

from pywfn.base import Mol
from pywfn.reader import LogReader
from pywfn.atomprop import activity
import matplotlib.pyplot as plt
import numpy as np
```

## 探究CHxFy的亲核亲电自由价

```python

root=rf"D:\BaiduSyncdisk\Articles\HFV\gfile\CHF亲核亲电\wfns"

nvals=[]
pvals=[]
for i in range(4):
    pathn=rf"{root}\CH{i}F{3-i}_wfn-.log"
    path0=rf"{root}\CH{i}F{3-i}_wfn0.log"
    pathp=rf"{root}\CH{i}F{3-i}_wfn+.log"

    moln=Mol(reader=LogReader(pathn))
    mol0=Mol(reader=LogReader(path0))
    molp=Mol(reader=LogReader(pathp))

    caler=activity.Calculator()
    caler.mols=[moln,mol0,molp]
    orders=caler.neFreeValence_v2(1)
    print(orders)
    nvals.append(float(orders[0,4]))
    pvals.append(float(orders[0,5]))
plt.xticks(range(4),[f'$CH_{i}F_{3-i}$' for i in range(4)])
plt.plot(pvals,marker='.',color='r')
plt.plot(nvals,marker='.',color='b')
```
