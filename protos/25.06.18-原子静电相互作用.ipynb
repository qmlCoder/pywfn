计算分子中两个原子之间的静电相互作用

$$
\rho ^{tot}(r)=\rho (r)+\rho ^{nuc}(r)
$$

电子与电子之间的排斥、核与核之间的排斥，电子与核之间的吸引

```python
import sys;sys.path.append('d:/code/pywfn')
import numpy as np
```

```python
from pywfn.base import Mole
from pywfn.reader import LogReader
from pywfn.gridprop import density,dftgrid
from pywfn.maths import rlib
```

```python
mole=Mole(LogReader(rf"d:\gfile\CO2.out"))
gridCaler=dftgrid.Calculator(mole)
grids,weits=gridCaler.molGrid()
print('格点数量：',grids.shape[0])
densCaler=density.Calculator(mole)
rhoA=densCaler.atmDens(grids,[1]).flatten()
rhoB=densCaler.atmDens(grids,[2]).flatten()
```

```python
atms=[8,6]
xyzs=[]
for i in [0,1]:
    x,y,z=[e.item() for e in mole.coords[i]]
    xyzs.append([x,y,z])
```

```python
res=rlib.calc_EIEBA_rs(atms,xyzs,rhoA,rhoB,grids,weits)
res
```
