```python
from pywfn.base import Mole
from pywfn.reader import LogReader
from pywfn.bondprop import order
import matplotlib.pyplot as plt
import numpy as np
```

```python
root=rf"d:\gfile\金属芳香性\rotated_structures\rotated_structures"
vals=[]
for i in range(36):
    path=rf"{root}\Ga-Ga-{i*10:0>3}.log"
    mole=Mole(LogReader(path))
    caler=order.Calculator(mole)
    deco,omat=caler.pi_pocv()
    val=omat[20,21]
    vals.append(val)
plt.plot(vals)
```

```python
angs=np.linspace(0, 2*np.pi, 36,endpoint=False)
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angs, vals, color='steelblue', linewidth=2)  # 边线
```
