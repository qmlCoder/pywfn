把轨道的每一个系数都当成向量进行投影

```python
from pywfn.base import Mole
from pywfn.reader import LogReader,FchReader
from pywfn.atomprop import direction
import numpy as np

mol=Mole(FchReader(rf"d:\gfile\pywfn\C6H6.fch"))

dirCaler=direction.Calculator(mol)

nato,nobt=mol.CM.shape

atoSyms=mol.coefs.atoSyms('raw')
atoAtms=mol.coefs.atoAtms('raw')
CMp=np.zeros_like(mol.CM)
for o in range(nobt):
    for a in range(nato):
        print(atoSyms[a])
        coef=mol.CM[o,a].item()
        match atoSyms[a]:
            case 'S':
                CMp[a,o]=coef
                continue
            case 'PX':
                vec=(coef,0,0)
            case 'PY':
                vec=(0,coef,0)
            case 'PZ':
                vec=(0,0,coef)
            case 'XX':
                vec=(coef*2,0,0)
            case 'YY':
                vec=(0,coef*2,0)
            case 'ZZ':
                vec=(0,0,coef*2)
            case 'XY':
                vec=(coef,coef,0)
            case 'XZ':
                vec=(coef,0,coef)
            case 'YZ':
                vec=(0,coef,coef)
        iatm=atoAtms[a] # 原子序号
        normal=dirCaler.normal(iatm)
        # print('normal',normal)
        # print('vec',vec)
        if normal is not None:
            prj=np.dot(vec,normal)*normal
    break
```
