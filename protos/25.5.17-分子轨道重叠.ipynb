使用数值方法实现计算两个片段的分子轨道之间的重叠

```python
import sys;sys.path.append('d:/code/pywfn')
```

```python
from pywfn.base import Mole
from pywfn.reader import LogReader
from pywfn.gridprop import wfnfunc,dftgrid
from pywfn.maths import rlib
import numpy as np
```

```python
mol=Mole(LogReader(rf"d:\gfile\pywfn\NH3BH3\NH3BH3.out"))
```

```python
grids,weits=dftgrid.Calculator(mol).molGrid() ## 获取分子的格点权重
```

```python
molN=Mole(LogReader(rf"d:\gfile\pywfn\NH3BH3\NH3.out"))
molB=Mole(LogReader(rf"d:\gfile\pywfn\NH3BH3\BH3.out"))
```

```python
wfnN=wfnfunc.Calculator(molN).obtWfns(grids,[4]).flatten()
wfnB=wfnfunc.Calculator(molB).obtWfns(grids,[4]).flatten()
np.sum(wfnN*wfnB*weits)
```

```python
## 获取两个原子的基函数索引
atos1=[]
atos2=[]
for i,atm in enumerate(mol.atoAtms):
    # print(atm)
    if atm==1:atos1.append(i)
    if atm==5:atos2.append(i)
atos1,atos2
```

```python

wfns0=wfnfunc.Calculator(mol).atoWfns(grids,0)[0] ## 所有基函数在相应格点的权重
wfns0
```

```python
for i in range(10):
    vals1=(mol.CM[atos1,i].reshape(-1,1)*wfns0[atos1,:]).sum(axis=0)
    vals2=(mol.CM[atos2,i].reshape(-1,1)*wfns0[atos2,:]).sum(axis=0)
    overlap=(vals1*vals2*weits).sum()
    print(i+1,overlap)
```
