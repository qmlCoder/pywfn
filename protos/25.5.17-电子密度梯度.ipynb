```python
from pywfn.base import Mole
from pywfn.reader import LogReader
from pywfn.gridprop import density,CubeGrid,dftgrid
import numpy as np
```

```python
mol=Mole(LogReader(rf"d:\gfile\pywfn\C6H6.out"))
# mol=Mole(LogReader(rf"d:\code\wmview_api\gfile\r3-pop.log"))
caler=density.Calculator(mol)
p0,p1=mol.spaceBorder
print(p0)
print(p1)
# shape,grids=CubeGrid().set_v1(p0,p1,0.2,4).get()
grids,weits=dftgrid.Calculator(mol).molGrid()
molDens0,molDens1,_=caler.molDens(grids,1)
```

```python
np.sum(molDens0*weits)
```

```python
ngrid=grids.shape[0]
lens=np.linalg.norm(molDens1,axis=1)
ratiox=np.divide(molDens1[:,0]**2,lens**2,out=np.zeros(ngrid),where=(lens!=0))
ratioy=np.divide(molDens1[:,1]**2,lens**2,out=np.zeros(ngrid),where=(lens!=0))
ratioz=np.divide(molDens1[:,2]**2,lens**2,out=np.zeros(ngrid),where=(lens!=0))
```

```python
elex=np.sum(molDens0*ratiox*weits)
eley=np.sum(molDens0*ratioy*weits)
elez=np.sum(molDens0*ratioz*weits)
elex,eley,elez,elex+eley+elez
```
