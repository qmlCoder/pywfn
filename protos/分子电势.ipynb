```python
import sys
sys.path.append('d:/code/pywfn')
from pywfn.base import Mol
from pywfn.reader import LogReader
from pywfn.spaceProp import potential
from pywfn.maths import cubeGrid
import numpy as np
import  matplotlib.pyplot as plt
```

```python
path="D:\gfile\pywfn\HCCH.out"
path="D:\gfile\pywfn\dingerxi.out"
mol=Mol(LogReader(path))
```

```python
caler=potential.Calculator(mol)
size,grid=cubeGrid(
    p0=np.min(mol.coords,axis=0),
    p1=np.max(mol.coords,axis=0),
    step=0.1,
    bord=1
)
caler.set_grid(grid)
```

```python
res=caler.nucPotential()
```

```python
mat=res.reshape(*size)
```
