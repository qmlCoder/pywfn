```python
import sys
sys.path.append('d:/code/pywfn')
from pywfn.base import Mol
from pywfn.reader import LogReader
from pywfn.spaceProp import wfnfunc
from pywfn.maths import cubeGrid
import numpy as np
import matplotlib.pyplot as plt
```

```python
path="D:\gfile\pywfn\C6H6.log"
reader=LogReader(path)
mol=Mol(reader)
caler=wfnfunc.Calculator(mol)
```

```python
R=5
p0=np.array([-1.0,-1.0,-1.0])*R
p1=np.array([1.0,1.0,1.0])*R
size,grid=cubeGrid(p0,p1,0.2)
grid.shape,size
```

```python
caler.set_grid(grid)
wfn=caler.obtWfn(21).reshape(size)
```

```python
plt.matshow(wfn[:,:,10],vmin=-0.05,vmax=0.05)
```

```python
mat=wfn.sum(axis=2)
print(mat.min(),mat.max())
plt.matshow(wfn.sum(axis=2),vmin=-0.05,vmax=0.05)
```
