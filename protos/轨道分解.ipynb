```python
from sympy import symbols, expand, Matrix, latex
from sympy.core.add import Add

# 定义符号变量
x, y, z = symbols('x y z')
T = symbols('T11 T12 T13 T21 T22 T23 T31 T32 T33')

# 定义变换后的坐标 (r' = r·T)
xp = T[0]*x + T[3]*y + T[6]*z  # x' = T11*x + T21*y + T31*z
yp = T[1]*x + T[4]*y + T[7]*z  # y' = T12*x + T22*y + T32*z
zp = T[2]*x + T[5]*y + T[8]*z  # z' = T13*x + T23*y + T33*z

# 定义基态 (B) 和激发态 (E) 的基函数列表
basis_B = [
    x, y, z,          # x, y, z
]

basis_E = [
    xp, yp, zp,          # x', y', z'
]

# 初始化系数矩阵
M = []

# 对每个激发态基函数展开并提取基态基函数的系数
for term_E in basis_E:
    expanded_term:Add = expand(term_E)
    row = [expanded_term.coeff(term_B) for term_B in basis_B]
    M.append(row)

# 转换为矩阵并转置（基态系数在行，激发态在列）
M = Matrix(M).T
M
```

```python
from sympy import symbols, expand, Matrix, latex
from sympy.core.add import Add

# 定义符号变量
x, y, z = symbols('x y z')
T = symbols('T11 T12 T13 T21 T22 T23 T31 T32 T33')

# 定义变换后的坐标 (r' = r·T)
xp = T[0]*x + T[3]*y + T[6]*z  # x' = T11*x + T21*y + T31*z
yp = T[1]*x + T[4]*y + T[7]*z  # y' = T12*x + T22*y + T32*z
zp = T[2]*x + T[5]*y + T[8]*z  # z' = T13*x + T23*y + T33*z

# 定义基态 (B) 和激发态 (E) 的基函数列表
basis_B = [
    x**3, y**3, z**3,          # xxx, yyy, zzz
    x*y**2, x**2*y, x**2*z,    # xyy, xxy, xxz
    x*z**2, y*z**2, y**2*z,    # xzz, yzz, yyz
    x*y*z                       # xyz
]

basis_E = [
    xp**3, yp**3, zp**3,           # x'^3, y'^3, z'^3
    xp*yp**2, xp**2*yp, xp**2*zp,  # x'y'^2, x'^2y', x'^2z'
    xp*zp**2, yp*zp**2, yp**2*zp,  # x'z'^2, y'z'^2, y'^2z'
    xp*yp*zp                        # x'y'z'
]

# 初始化系数矩阵
M = []

# 对每个激发态基函数展开并提取基态基函数的系数
for term_E in basis_E:
    expanded_term:Add = expand(term_E)
    row = [expanded_term.coeff(term_B) for term_B in basis_B]
    M.append(row)

# 转换为矩阵并转置（基态系数在行，激发态在列）
MF = Matrix(M).T
MF
```

```python
for i in range(10):
    for j in range(10):
        print(f'M[{i}][{j}]={MF[i,j]}')
```
